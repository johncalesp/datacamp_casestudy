## 7. MODEL DEVELOPMENT <a name="model"></a>

```python
import catboost as cb
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import os
sns.set()
```

```python
!jupyter nbextension enable --py widgetsnbextension
```

## Modeling
#### I will use the package CatBoost. I have been using this package for some of my projects. It is based on gradient boosting similar to XGboost and provides a more friendly support for categorical data

### Preparing the data in features and target value

```python
X_df = bmw_filtered_eng_mpg.drop("price", axis=1)
X_df.info()
```

```python
y = bmw_filtered_eng_mpg["price"]
y[:5]
```

#### Selecting the categorical features

```python
cat_features = [0,2,4] # the categorical features are Model, Transmission and Fuel type
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size = 0.2, random_state = 42)
```

#### Generating an input format for the model using the "Pool" class

```python
train_dataset = cb.Pool(X_train, y_train, cat_features = cat_features ) 
test_dataset = cb.Pool(X_test, y_test, cat_features = cat_features )
```

```python
model_grid = cb.CatBoostRegressor(loss_function='RMSE',
                        logging_level='Silent')
```

```python
#Best so far :
# iterations : 500
# lr: 0.15
#depth: 8
# l2_leaf_reg: 0.2
grid = {'iterations': [400,500,600,700],
        'learning_rate': [0.10, 0.15, 0.25],
        'depth': [6,8,10],
        'l2_leaf_reg': [0.2, 0.4, 0.6]}
model_grid.grid_search(grid, train_dataset, verbose=False)
```

```python
model_grid.score(X_test, y_test)
```

#### Generating a new model using the best features

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
```

```python
train_dataset = cb.Pool(X_train, y_train, cat_features = cat_features ) 
valid_dataset = cb.Pool(X_valid, y_valid, cat_features = cat_features ) 
test_dataset = cb.Pool(X_test, y_test, cat_features = cat_features )
```

```python
model_set_param = cb.CatBoostRegressor(loss_function='RMSE', iterations= 500,
        learning_rate= 0.15,
        depth= 8,
        l2_leaf_reg= 0.2,
        logging_level='Silent')
```

```python
model_set_param.fit(train_dataset, eval_set=valid_dataset, plot=True)
```

```python
model_set_param.score(X_test, y_test)
```

```python
model_set_param.get_feature_importance(prettified=True)
```

#### Plotting Features Importance

```python
plt.figure(figsize=(12, 6));
ax = sns.barplot(x="Importances", y="Feature Id", data=model_set_param.get_feature_importance(prettified=True));
ax.set_title('CatBoost features importance:', fontsize=16);
ax.set_xlabel('Importance', fontsize=14);
ax.set_ylabel('Feature', fontsize=14);
plt.show()
```

#### Reviewing features importance using Shap values

```python
import shap
explainer = shap.TreeExplainer(model_set_param)
shap_values = explainer.shap_values(train_dataset) 
shap.summary_plot(shap_values, X_train)
```

```python
pred = model_set_param.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, pred)))
r2 = r2_score(y_test, pred)
print('Testing performance')
print('RMSE: {:.2f}'.format(rmse))
print('R2: {:.2f}'.format(r2))
```

#### The model presents a 96% explaintability for the price and predicts it's value with a +/- 2240.49 (pounds) precision
